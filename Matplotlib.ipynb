5. Vẽ trên nhiều biểu đồ sử dụng vòng lặp

```python
import numpy as np
from numpy.random import seed, randint
```

```python
fig, axe3 = plt.subplots(1,2, figsize=(10,3), sharey=True, dpi=120)
#dinh nghia mau
colors={0:'g',1:'b',2:'r',3:'y'}
markers={0:'o',1:'x',2:'*',3:'p'}
#plot cho moi axe3 
for i, ax in enumerate(axe3.ravel()):
    ax.plot(sorted(randint(0,10,10)),sorted(randint(0,10,10)),marker=markers[i],color=colors[i])
    ax.set_title('Ax: '+ str(i))
    ax.yaxis.set_ticks_position('none')
plt.suptitle('for subplots in onr figure', verticalalignment='bottom',fontsize=12)
plt.tight_layout()
```

6. Điều chỉnh view trục x và y

```python
from matplotlib.ticker import FuncFormatter
```

```python
#ve 1 bieu do don gian
plt.figure(figsize=(7,5), dpi=100)
X = np.linspace(0,2*np.pi, 100) #Trả về các số cách đều nhau trong một khoảng cụ thể. 
                                # Trả về `num` các mẫu cách đều nhau, được tính trên khoảng [`bắt đầu`,` dừng`]
                                # Có thể tùy chọn loại trừ điểm cuối của khoảng thời gian.
plt.plot(X,np.sin(X))
plt.plot(X,np.cos(X))
# dieu chinh truc x va y
plt.xticks(ticks=np.arange(0,361/57.2985,90/57.2965),label=[r'0',r'2',r'5'],fontsize=12,rotation=30)#rotation : góc lệch của text so với trục thẳng đứng (trục y)
#1 radian = 57.2985 degrees
plt.yticks(ticks=np.arange(-1,1,0.2),fontsize=12,rotation=30)#rotation : góc lệch của text so với trục thẳng đứng (trục y)
# convert radian sang do
def rad_to_degrees(x, pos):
    'converts radians to degrees'
    return round(x*57.2985,1)# 1 la tra ve 1 so sau dau phay
formatter = FuncFormatter(rad_to_degrees)
plt.gca().xaxis.set_major_formatter(formatter)#Phương thức plt.gca() trả về đối tượng Axes hiện tại, tức là biểu đồ hiện tại.

plt.grid(linestyle='--',linewidth=3,alpha=0.1)#linewidth=0.5,alpha=0.1 là nền của hình
plt.title('sine and cosine waves',fontsize=14)
plt.show()
```

### projext data visualization - california cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('california_cities.csv')
cities.head()
cities['area_total_km2'].sort_values(ascending=False)
```

```python
#extract Latd and Longd

lat,lon = cities['latd'],cities['longd']
population,area = cities['population_total'],cities['area_total_km2']
plt.figure(figsize=(8,6))
plt.style.use('seaborn')

plt.scatter(lon,lat,
           c=np.log10(population),cmap='viridis',s=area,linewidth=0,alpha=0.5);
plt.axis('equal');
plt.xlabel('lontitude')
plt.ylabel('lattitude')
plt.colorbar(label='log10(population')
plt.clim(3,7)

#create a legend for cities's size
area_range = [50,100,300,500]
for area in area_range:
    plt.scatter([],[],s=area,c='k',alpha=0.4,
               label=str(area)+'km$^2')
plt.legend(labelspacing=1.5,title='city area')

plt.title('California cities: population and area distribution');
```
